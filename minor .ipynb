# Countdown Timer


```python
import time
import tkinter as tk

##tkinter is used to create desktop application 

class CountdownTimer:    
    def __init__(self, master):  #init - for attribute of object(new var created for class )
        self.master = master
        master.title("Countdown Timer")

        self.time_label = tk.Label(master, font=("Helvetica", 48))
        self.time_label.pack(pady=20)

        self.start_button = tk.Button(master, text="Start", command=self.start_timer)
        self.start_button.pack(pady=10)

        self.reset_button = tk.Button(master, text="Reset", command=self.reset_timer, state=tk.DISABLED)
        self.reset_button.pack(pady=10)

        self.pause_button = tk.Button(master, text="Pause", command=self.pause_timer, state=tk.DISABLED)
        self.pause_button.pack(pady=10)

        self.resume_button = tk.Button(master, text="Resume", command=self.resume_timer, state=tk.DISABLED)
        self.resume_button.pack(pady=10)

        self.seconds_left = 0
        self.timer_running = False
        self.paused = False

    def start_timer(self):  #for activate timer
        self.seconds_left = 15 # given time 
        self.timer_running = True
        self.paused = False
        self.start_button.config(state=tk.DISABLED)
        self.reset_button.config(state=tk.NORMAL)
        self.pause_button.config(state=tk.NORMAL)
        self.resume_button.config(state=tk.DISABLED)
        self.countdown()

    def countdown(self):
        while self.seconds_left > 0 and self.timer_running:
            if not self.paused:
                self.time_label.config(text=self.seconds_left)
                self.master.update()
                time.sleep(1)
                self.seconds_left -= 1
        self.timer_running = False
        self.start_button.config(state=tk.NORMAL)
        self.reset_button.config(state=tk.DISABLED)
        self.pause_button.config(state=tk.DISABLED)
        self.resume_button.config(state=tk.DISABLED)

    def reset_timer(self): #reset fn
        self.timer_running = False
        self.paused = False
        self.seconds_left = 0
        self.time_label.config(text="")
        self.start_button.config(state=tk.NORMAL)
        self.reset_button.config(state=tk.DISABLED)
        self.pause_button.config(state=tk.DISABLED)
        self.resume_button.config(state=tk.DISABLED)

    def pause_timer(self):  #pause fn
        self.paused = True
        self.resume_button.config(state=tk.NORMAL)
        self.pause_button.config(state=tk.DISABLED)

    def resume_timer(self):  #resume fn
        self.paused = False
        self.resume_button.config(state=tk.DISABLED)
        self.pause_button.config(state=tk.NORMAL)
        self.countdown()

root = tk.Tk()
my_timer = CountdownTimer(root)
root.mainloop()
```
